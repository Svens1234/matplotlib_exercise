```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(0,100)
b=a-5
c=a**3-b**2
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(a,b)
axes.set_xlabel('a label')
axes.set_ylabel('b label')
axes.set_title('title label')
```

```python
figure=plt.figure()
axes1=figure.add_axes([0,0,1,1])
axes2=figure.add_axes([0.2,0.5,0.2,0.3])
axes1.plot(a,b)
axes2.plot(b,c)
```
